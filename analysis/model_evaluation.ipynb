```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
import chardet

def read_text(file_path):
    with open(file_path, 'rb') as f:
        raw_data = f.read()

    for encoding in ['utf-8', 'ISO-8859-1', 'cp1252']:
        try:
            return raw_data.decode(encoding)
        except UnicodeDecodeError:
            continue

    raise UnicodeDecodeError(f"Failed to decode {file_path} with common encodings.")


```

```python
scores = []
for idx, row in test_data.iterrows():
    resume_text = read_text(f"../test_data/resumes/{row['resume_file']}")
    jd_text = read_text(f"../test_data/jds/{row['jd_file']}")

    score = util.cos_sim(model.encode(resume_text), model.encode(jd_text)).item()
    scores.append(score)
```
